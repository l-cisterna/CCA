### 🎡 For Loops () 


---






La idea es **hacer algo de forma iterada (repetida)**. Es decir repetir lo mismo, pero tal vez cambiando algo en cada iteración. 

<br>

El for loop se arma de la siguiente manera: 

**for nombre_cualquiera in secuencia: Blocke Indentado**

````python
#secuencia
seq = [1,2,3,4,5]

for x in seq:
    if x>3:
        print(x+32)
    else:
        print(x)
````

````python
#no es necesario hacer algo relacionado a la variable
for item in seq:
    print(item)
````

**zip**  es una función que permite hacer for loop con dos variables a la vez.

````python
nombres = ["Khun", "Vampiro", "Messi"]
habilidad_futbol = [ 6, 9.8, 6.3 ]

for n, h in zip(nombres, habilidad_futbol):
  print("{}, tiene una habilidad en el futbol de: {} ".format(n,h))
````

````python
nombres = ["Vampiro", "Khun", "Messi"]
habilidad_futbol = [ 6, 9.8, 6.3]

for n, h in zip(nombres, habilidad_futbol):
  print(n + ", tiene una habilidad en el futbol de: " + str(h))
````


## 🌌  Métodos Summary
Se trata de métodos para encontrar los máximos y mínimos dentro de un array, así como el lugar donde están ubicados


````python
# Seed para mantener el mismo resultado
np.random.seed(100)

# Matriz Aleatoria Normal
A = np.random.rand(3,3)
A
````

````python
print(A)

print("\n Máximo: ")
print(A.max()) #Máximo de toda la array

print("\n Mínimo")
print(A.min()) #Mínimo de toda la array

print("\n Indice del Máximo por columna")
print(A.argmax(axis=0)) # índice del máximo, por columna

print("\n Inidece del Mínimo por columna")
print(A.argmin(axis=0)) # índice del mínimo, por columna
````

````python
print("\n Media:")
print(A.mean()) #Media general

print("\n Media por columna:")
print(A.mean(axis=0)) #Media por columna

print("\n Desvío General:")
print(A.std())

print("\n Desvío por columna:")
print(A.std(axis=1))
````
